```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
all_df = []
```

```python
train_size = [1453, 1000, 700, 350, 175, 100]
```

```python
for ii in range(6):
    results = pd.read_csv(f'resnet_{ii + 1}_eval.csv')
    results['train_size'] = train_size[ii]
    all_df.append(results)
```

```python
data = pd.concat(all_df)
```

```python
ax = sns.regplot(x="train_size", y='Value', data=data[(data["Metric"]=="r2")], x_estimator=np.mean)
```

```python
ax = sns.regplot(x="train_size", y='Value', data=data[(data["Metric"]=="mae")], x_estimator=np.mean)
```
